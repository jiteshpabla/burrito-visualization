```
import pandas as pd
import string
```

```
df = pd.read_csv("Burrito.csv")
df
```

```
df1 = df[['Location','Burrito', "Date", "Cost", "Hunger", "Tortilla", "Temp", "Meat", "Fillings", "Uniformity", "Salsa", "Synergy", "Wrap", "overall"]]
```

```
df1.dropna(inplace=True)
df1["Location"] = df1['Location'].str.replace('[{}]'.format(string.punctuation), '')
df1["Location"] = df1["Location"].str.title()
df1
```

```
df1.Location.nunique()
```

```
# get all restraurants with more than 10 reviews
df2 = df1.groupby(['Location']).filter(lambda x: len(x) > 5)
df2
```

```
df2.groupby(['Burrito']).count()
```

```
df_coolVals = df2.groupby(['Location']).mean()
df_coolVals
```

```
df_coolVals.to_csv("data1.csv")
```
